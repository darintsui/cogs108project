# Research Question

How has COVID-19 and its trend of total cases and total deaths impacted mental health, as alluded to by Google search data from various countries?

# Dataset(s)

For the following datasets, we have specified which specific files we are using in parentheses next to the Dataset Names

1st Dataset
- Dataset Name: COVID-19 and Mental Health Search Terms (“search_term_canada.xlxs,” “search_term_iran.xlxs,” “search_term_italy.xlxs,” “search_term_japan.xlxs,” “search_term_sk.xlsx,” “search_term_uk.xlsx,” “search_term_us.xlsx”)
- Link to the dataset: https://www.kaggle.com/luckybro/mental-health-search-term (each dropdown has a country [for Canada’s dataset, it’s the first one])
- Number of observations: 51 rows of weekly google trends of mental health search terms for each country; Total observations: 357 
This dataset compiles data from Google Trends, which gives an arbitrary score from 1 to 100 on the amount of times the keyword is searched weekly by country. This data set tracks keywords related to mental health such as “depression,” “anxiety,” “obsessive compulsive disorder,” etc.

2nd Dataset 
- Dataset Name: COVID-19 data from John Hopkins University (“CONVENIENT_global_confirmed_cases.csv” and “CONVENIENT_global_deaths.csv” datasets)
- Link to the dataset: https://www.kaggle.com/antgoldbloom/covid19-data-from-john-hopkins-university 
- Number of observations: 385 rows of the total number of cases/deaths reported each day (updated daily)
Under the convenient global confirmed cases dataset, the author of this dataset has already cleaned up data given from Johns Hopkins University. The data’s columns are the countries, and each row is a different day, so the values in the data set represent how many new confirmed Covid cases a country has for a particular day.. The convenient global deaths dataset follows a similar format as well and reports daily Covid deaths per country. 


Plan To Combine Datasets: 
- We plan to combine these datasets by exploring the correlation between the score (which determines the popularity of the word searched by week per country) assigned to a certain mental health keyword and the total number of cases and deaths in each country. For example, if there’s a record high number of COVID cases a certain week in the US (say the 100,000 milestone), we would reason that the number of keywords searched relating to mental health illness would be high that week.

Starting with the global_cases and global_death dataframes, we want to just look at the shape and the last few rows.

So the shapes and formats of the global_cases and global_death match which is good. However, there are a lot of countries represented in these two datasets. Since we only have search term data for a handful of countries, we won't be able to use all of the countries in these Covid Cases and Deaths datasets. So the next step in our cleaning process is to drop the countries in those data sets that we do not have search term data for.

```python
#Drop countries in global_cases and new_global_death that aren't included in countries with mental health data
new_global_cases = global_cases[['Country/Region', 'Canada', 'Iran', 'Japan', 'Korea, South', 'United Kingdom', 'US']]
new_global_death = global_death[['Country/Region', 'Canada', 'Iran', 'Japan', 'Korea, South', 'United Kingdom', 'US']]
```

Next we want to adjust the column names slightly since the first column contains the dates of interest rather than the country names. We also want to remove any rows that have NaN values. For example, the row Province/State has NaN values since not all countries are separated into such regions.

```python
#Rename dataset to make the column name more logical
new_global_cases = new_global_cases.rename(columns={'Country/Region': 'Date'})
new_global_death = new_global_death.rename(columns={'Country/Region': 'Date'})
```

```python
#check and remove nan values
new_global_cases.isnull().sum()
new_global_cases[new_global_cases.isnull().any(axis = 1)]
```

```python
new_global_death.isnull().sum()
new_global_death[new_global_death.isnull().any(axis = 1)]
```

new_global_cases and new_global_death have their first row containing null values of Province and State. We won't be analyzing province and state of the data, so we can safely remove this row.

```python
#drop nas
new_global_cases = new_global_cases.dropna()
new_global_death = new_global_death.dropna()
```

At the same time, since we only have search term data with dates up to 5/31/20, we won't be able to use all of the date data on the Covid Cases and Deaths datasets. So the next step in our cleaning process is to drop the datas in those data sets that we do not have search term data for.

```python
#all of the search data ends at 5/31/2020, thus we drop all the rows after 5/31/2020 in the new_global_cases datafram
count = new_global_cases.index[new_global_cases['Date'] == '5/31/20']
new_global_cases = new_global_cases[:130]
new_global_death = new_global_death[:130]
```

Everything looks good now for the covid cases and deaths datasets as they are all the same shape. Now we can move on to the Mental Health search term data sets for the different countries. These data sets are already very tidy. All of them are the same size (51 x 10) and they all have the same columns and dates. Also, all of the column titles are descriptive and there are no NaN values. One thing we can do though is to condense the 7 country data sets into one. This way we have less dataframes to work through and it would mimic the covid cases and death dataframes which have multiple countries in one set.

```python
#Merge all the data of mental health search
canada_search['country'] = 'Canada'
iran_search['country'] = 'Iran'
italy_search['country'] = 'Italy'
japan_search['country'] = 'Japan'
sk_search['country'] = 'South Korea'
uk_search['country'] = 'United Kindom'
us_search['country'] = 'United States'
merge = [canada_search, iran_search, italy_search, japan_search, sk_search, uk_search, us_search]
global_mental_health = pd.concat(merge)
global_mental_health
```

Everything looks good to go now for EDA.
